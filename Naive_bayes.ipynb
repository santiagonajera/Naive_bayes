```python
#Tenemos datos de clientes anteriores con sus edades y montos de sus últimas
#compras, así como si compraron (1) o no compraron (0) el producto.
#Entrenamos un modelo de Naive Bayes con estos datos y luego usamos el modelo
#para predecir si un nuevo cliente comprará el producto o no,
#y con qué probabilidad.

from sklearn.naive_bayes import GaussianNB
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo actualizados (historial de compras)
# X representa las características: [edad, monto de la última compra]
# y representa las etiquetas (0: No compró, 1: Compró)
X = np.array([[25, 500], [30, 1200], [35, 900], [40, 1100], [45, 1500]])
y = np.array([1, 0, 1, 0, 1])

# Crear el modelo Naive Bayes
model = GaussianNB()

# Entrenar el modelo
model.fit(X, y)

# Nuevos datos para predecir
nuevo_cliente = np.array([[33, 1300]])

# Realizar la predicción
prediccion = model.predict(nuevo_cliente)
probabilidad = model.predict_proba(nuevo_cliente)

print("Predicción (0: No comprará, 1: Comprará):", prediccion)
print("Probabilidad de cada clase:", probabilidad)

# Crear un gráfico para visualizar las probabilidades
prob_clases = model.predict_proba(X)

plt.scatter(X[:, 1], y, color='green', label='Datos reales')
plt.scatter(nuevo_cliente[:, 1], prediccion, color='blue',
            label='Nueva predicción')
plt.xlabel('Monto de la última compra')
plt.ylabel('(0: No compró, 1: Compró)')
plt.legend()
plt.title('Predicción de compra basada en monto de la última compra y edad')
plt.show()
```
